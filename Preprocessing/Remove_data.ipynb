```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv', index_col=0)
df
```

```python
mask = df['Description'].str.len()<1000
df = df[mask].reset_index(drop=True)
df
```

```python
df = df[df['Genre']!='Best Seller'].reset_index(drop=True)
df
```

```python
filter = df[df['Genre'] == 'Tâm Lý - Kỹ Năng Sống'].copy()
filter
```

```python
df.drop(filter.head(600).index, inplace=True)
```

```python
df = df.reset_index(drop=True)
df
```
